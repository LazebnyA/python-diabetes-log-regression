Distributions

```python
def plot_categorical_distribution(data, column_name):
    value_counts = data[column_name].value_counts()

    value_counts.plot.bar()

    plt.xlabel(column_name)
    plt.ylabel('Frequency')
    plt.title(f'Distribution of {column_name}')
    plt.show()
```

```python
categorical_variables = [
    "Diabetes_binary",
    "HighBP",
    "HighChol",
    "CholCheck",
    "Smoker",
    "Stroke",
    "HeartDiseaseorAttack",
    "PhysActivity",
    "Fruits",
    "Veggies",
    "HvyAlcoholConsump",
    "AnyHealthcare",
    "NoDocbcCost",
    "DiffWalk",
    "Sex",
    "Age",
    "Education",
    "Income",
    "GenHlth"
]

df[categorical_variables] = df[categorical_variables].apply(lambda data: data.astype(int))
df.head()

df.to_csv('prepared_data.csv')
```
